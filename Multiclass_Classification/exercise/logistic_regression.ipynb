```python
# Import Libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Load iris
iris = load_iris()
dir(iris)
```

```python
# The Iris dataset contains measurements of 150 iris flowers from three species: setosa, versicolor, and virginica. 
# Each flower has four features: sepal length, sepal width, petal length, and petal width (all in cm). The dataset includes:
# data: 150 samples Ã— 4 features (flower measurements).
# target: Labels for species (0 = setosa, 1 = versicolor, 2 = virginica).
# target_names: Names of the three species.
# feature_names: Names of the four features.
# DESCR: Full description of the dataset.
iris['data']
```

```python
# Target refers to Labels for species (0 = setosa, 1 = versicolor, 2 = virginica).
iris.target[0:5]
```

```python
iris.target_names
```

```python
# Split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
len(X_train)
len(X_test)
```

```python
# Call the Logistic Regression dataset
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
# What is the score of the model
model.score(X_test, y_test)
```

```python
# What is the label of 67
iris.target[67]
```

```python
# Lets predict this image 67
model.predict([iris.data[67]])
```

```python
# Lets predict the iris
model.predict(iris.data[0:5])
```

```python
# Lets see which flower are not being predicted well
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
# We can see it better here
# For example all flowers are being represented well
import seaborn as sn
plt.Figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
